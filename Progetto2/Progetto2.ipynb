```python
import os.path
from zipfile import ZipFile

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
zip_name = "air+quality.zip"
zip_link = "https://archive.ics.uci.edu/dataset/360/air+quality"
csv_name = "AirQualityUCI.csv"
```

Download del dataset

```python
if not os.path.exists(zip_name):
    print("Error: Dataset not found, download id from " + zip_link)
else:
    with ZipFile(zip_name) as f:
        f.extractall()
```

```python
with open(csv_name) as dataFile:
    data_raw = pd.read_csv(dataFile, sep=",")
```
